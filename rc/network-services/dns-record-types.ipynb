***
###### DNS Record Types
***

Accessing domain name services server in order to gather information about the name or the IP address of a resource, we’re really looking through a local database that has all of these associations between a fully qualified domain name and an IP address.

These records on a DNS server are called the resource records (RR). 

Its really a text file that happens to contain all of the information we need to be able to perform these lookups. 

There are over 30 different types that you may find inside of a DNS server.

< [Table of Contents](#top) | [References](#references) >
<a id="DNSServerDatabaseTextFile"></a>
***
###### DNS Server Database Text File
***

The database of a DNS server is simply a text file. 

Would configure the DNS server and this is where all the lookups will take place. 

Can see information in this sample lookup file that has information about the DNS server itself and information associated with the caching. 

Also see mail exchange records. 

There’s records that are specifying the name and the IP address of other devices.

This lookup also contains aliases - allows us to use different names to associate to a single fully qualified domain name.

***
###### Associate Fully Qualified Domain Name (FQDN) to an IP (Internet Protocol)
***

One of the most important roles of a DNS is the one where we are associating an IP address with a fully qualified domain name. 

This section of the lookup file, can see there are some machine names and IP addresses associated with those machine names.

< [Table of Contents](#top) | [References](#references) >
<a id="AddressRecordsAaaaa"></a>
***
###### Address Records ``A`` or ``AAAA``
***

These address records are called ``A`` records or quad ``AAAA`` records. 

Used to associate the name with an IP address. 

Use the ``A`` record to associate an IPv4 address with the name of a device.

Use quad ``AAAA`` record to associate an IPv6 address with the name of a device. 

Example of a record having the name ``www.professormesser.com``.

Specification of the class of this record - the class of this record will be the internet class ``IN``.

This is an ``A`` record, so this will be an IPv4 address.

Can see the IPv4 address is also included as part of this record that associates this IP address with ``www.professrmesser.com``.

Can also add a semi-colon at the end of the record to designate a comment, so that you can add other contexts to this particular record.

< [Table of Contents](#top) | [References](#references) >
<a id="CanonicalNameRecordCNAME"></a>
***
###### Canonical Name Record - ``CNAME``
***

May want to associate a number of different names with an existing device. 

In those situations, may want to use a canonical name record or ``CNAME`` record.

This is where you would add the alias to something that already exists. 

Had a device called ``mail.mydomain.name``, could also call this the ``ftp`` server for ``ftp.mydomain.name``, ``www.mydomain.name``. 

The DNS server will recognize that this is a canonical name - will associate it with the primary name of this device and then perform the proper lookup.

< [Table of Contents](#top) | [References](#references) >
<a id="ServiceRecordsSRV"></a>
***
###### Service Records - ``SRV``
***

There are also DNS records that can help find particular services on the network.

Could create a service record or ``SRV`` record that would allow Windows devices to find Windows domain controllers on the network or you might add a service record so that voiceover IP devices can find the centralized voiceover IP controller on the network.

Here’s the service record that you use for Windows devices to find an Active Directory controller.

Can see the service is ``\_ldap``.

The protocol name is ``tcp.domain.com.``.

The time to live (``TTL``) is 300 seconds.

This is the internet class (``IN``).

It is an ``SRV`` or service record.

There’s priority and weight values, a port number, and then the target that allows you to associate this particular service (``\_ldap.tcp.domain.com.``) with a particular device (``s1.domain.com``).

< [Table of Contents](#top) | [References](#references) >
<a id="MailExchangerRecordMX"></a>
***
###### Mail Exchanger Record - ``MX``
***

If in charge of the mail servers on the network, then you know the ``MX`` record or the mail exchange record is an extremely important record that allows third parties to be able to find your local mail servers.

In the DNS configuration file, the ``MX`` record starts with the ``IN`` class (for internet). 

It is an ``MX`` record and then you put the name of the mail server (``mail.mydomain.name``).

Later in the configuration, will find the name of that device is listed with an ``A`` record that will specify the IP address of that mail server.

< [Table of Contents](#top) | [References](#references) >
<a id="NameServerRecordNS"></a>
***
###### Name Server Record - ``NS``
***

Many DNS servers also contain configuration details for additional name servers on the domain.

These are ``NS`` records or name server records.

The format of a name server record would specify the class of this record, which is internet - ``IN``.

This is the name server record, or ``NS``.

Then will list out the names associated with those name servers (``ns1`.example.com`` - ``ns2.example.com``). 

Later on in the configuration, can find ``A`` (Address Records), that will associate the name of the name server to a specific IP address.

***

Commonly use DNS servers to associate a name with an IP address.

Can also do the reverse where we provide the name server with an IP address and it will provide us with the name.

<a id="PointerRecordPTR"></a>
***
###### Pointer Record - PTR
***

Pointer records, or ``PTR`` records, are effectively the reverse of an ``A`` or a quad ``A`` record.

In the configuration of our DNS, we would list out the IP addresses associated with these devices. 

Would specify that these are ``PTR`` or pointer records, and then we would provide the fully qualified domain name of these devices.

<a id="3rdPartyAccessingDNS"></a>
***
###### 3rd Party Accessing DNS
***

Another common DNS resource record found is a text record, or a ``TXT`` record. 

This record usually contains information that will be valuable to other third party services that are accessing your DNS.

< [Table of Contents](#top) | [References](#references) >
<a id="SenderPolicyFrameworkSPF"></a>
***
###### Sender Policy Framework - SPF
***

A text record (``TXT`` might contain ``SPF`` protocol information – this stands for Sender Policy Framework. 

Its a type of text message that helps prevent someone else from spoofing your domain name as emails are sent out across the internet. 

Other mail servers can check this text record and determine if this particular email address really came from your email server.

This ``SPF`` ``TXT`` record contains information that tells other mail servers what the valid mail servers will be for ``professormesser.com``.

The domain name, ``professormesser.com`` is listed.

Got a timeout value (``ttl``) of 300 seconds.

It is a text record (``TXT``).

The ``SPF`` information is ``v=spf1 include:mailgun.org ~all``.

Third parties will be able to look at this text record on the DNS server, and see that if mail is coming from ``mailgun.org``, than it is authorized as a mail server for ``professormesser.com``.

< [Table of Contents](#top) | [References](#references) >
<a id="DomainKeysIdentifiedMailDKIM"></a>
***
###### Domain Keys Identified Mail - DKIM
***

Can also use text records (``TXT``) to add ``DKIM``, or Domain Keys Identified Mail.

This allows mail servers to digitally sign outgoing mail.

When the external mail server receives this, they can confirm that this was, indeed, digitally signed by your mail server.

Effectively adding your public key into this particular DNS text record. 

The domain key text record listed here, shows the domain key information for ``professormesser.com``.

This is a text value and can see the entire public key is listed under this ``DKIM1`` value. 

Allows third-party mail servers that have received a digitally signed email from the mail server to get the public key and validate that signature.
