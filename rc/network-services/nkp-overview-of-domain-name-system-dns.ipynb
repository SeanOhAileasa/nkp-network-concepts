###### Overview of Domain Name System (DNS)
***

Converting from a human readable name like ``professormesser.com`` and accessing a website, there has to be some conversion to IP addresses.

The domain name system provides that conversion process.

DNS is a hierarchical system which follows a very specific path to be able to find exactly the server to locate and it’s a database that’s very distributed.

There are many DNS servers around the world - over ``13`` clusters of root servers at the core of DNS - often find generic top-level domains like a ``.com``, a ``.org``, or a ``.net``. 

There are also country code top-level domains for the US, Canada, UK, and other countries.

< [Table of Contents](#top) | [References](#references) >
<a id="DNSHierarchy"></a>
***
###### DNS Hierarchy
***

Example of the hierarchy that’s used with the DNS.

Very often will see a ``.com`` address or a ``.net``, and that’s the top-level domain of this hierarchy.

Underneath ``.com``, might have another domain name such as ``professormesser`` - can put together ``professormesser.com``.

Of course, at ``professormesser`` there may be a web server and that web server’s name may be ``www``.

There may also be a ``mail.professormesser.com``, an ``east.professormesser.com``, and a ``west.professormesser.com``.

Can have servers that were one level down from that - may be an ``ethan.west.professormesser.com`` and so on.

<a id="FullyQualifiedDomainNameFQDN"></a>
***
###### Fully Qualified Domain Name - FQDN
***

Visually, can see how this domain name might work. 

Have a Fully Qualified Domain Name (FQDN) that encompasses everything within ``professormesser.com``.

Within, there may be the web server for ``www.professormesser.com``. 

May break out these domains even further to have a ``west.professormesser.com`` and an ``east.professormesser.com``.

There may be servers that exist in these different regions, and can specify their FQDN as ``ethan.west.professormesser.com`` and ``katie.east.professormesser.com``.

###### DNS Process
***

DNS is a very distributed database and to be able to find all of those different services that are located as those fully qualified domain names, may need to ask a number of servers before we’re able to fully resolve their IP addresses. 

Start with a laptop - refer to this device as a resolver.

This is a device that will, for instance, be trying to find the IP address that might be associated with the ``www.professormesser.com``.

This device is configured for a local name server that it will reference.

This might be a local server that’s within an organization or it might be a third-party name server that you’re using.

This resolver will query that name server to ask where is the IP address for ``www.professormesser.com``.

The first request will be made from the resolver to the local name server asking if this particular device knows where ``www.professormesser.com`` happens to be.

The local name server does not have this IP address in its cache, so it needs to start asking other name servers where it can go to find the ``www.professormesser.com`` IP address.

Its first step is going to be talking to this root server.

The root server is going to respond back and say if looking for anything with a ``.com``, need to go to the ``.com`` name server.

So the next step will be the local name server will talk to the ``.com`` DNS server. 

The ``.com`` name server will have a list of where my local DNS server is for ``professormesser.com``, and it will provide that information to the local name server. 

The local name server will then query the ``professormesser.com`` name server, which obviously will have the IP address associated with ``www.professormesser.com`` and that information is provided to the local name server.

At this point, the local name server will let the resolver know what that IP address happens to be, and it will store that information in a cache. 

If anyone else that uses this local name server needs to access ``www.professormesser.com``, they can get that information from the local cache rather than going through all of these queries all over again.

<a id="InternalDNS"></a>
***
###### Internal DNS
***

Many organizations will maintain an internal DNS. 

There are internal servers that you need to have some type of reference to that you may not want to make public on an external DNS server. 

These internal DNSs are usually installed and maintained by your local network team, and it usually is going to have your internal information that no one else should have access to. 

It’s very common, for example, to run a DNS service on a Windows Server so that you’re able to keep up with all of the different Windows devices on your internal private network.

< [Table of Contents](#top) | [References](#references) >
<a id="ExternalDNS"></a>
***
###### External DNS
***

If an organization doesn’t need an internal DNS server, might want to use an external DNS.

A good example are the ones that are managed by Google on their ``Google DNS`` or ``Quad9``.

These external DNSs will obviously not have any information about any internal devices, but you’re able to use and maintain those external resources without having to run your own server internally in your organization.

< [Table of Contents](#top) | [References](#references) >
<a id="ThirdpartyDNS"></a>
***
###### Third-party DNS
***

There’s also a middle ground where you can have a third party providing you with your own internal DNS services. 

Especially if in a very large environment, may be having to manage many different DNS servers. 

Instead of you managing that, you might want to hand that off to a third party where you’ve outsourced your internal DNS function to this third party. 

All of these DNS servers are running in the cloud.

Might have multiple instances of them available around the world, and it would still provide you with the same internal functionality except you’re using a cloud-based service external to your organization.

This also might provide you with some capabilities you normally wouldn’t have if you were maintaining your own servers.

May have high availability to these DNS services available all the time. 

Ma be low latency servers providing very fast response time.

Need additional servers spun up, simply request that from the third-party DNS and you’re able to scale up and scale down as needed.
