***
###### Physical Network Map
***

Tasked with creating some documentation of your network configuration, might want to create a physical network map. 

Shows exactly what devices might be on the network, and it would show the physical cable connections from one interface to another.

It's also common for people to create a physical map of where devices might be installed in a particular rack that would allow you to go into a data centre to a specific rack and identify exactly the piece of equipment that's listed in your documentation. 

< [Table of Contents](#top) | [References](#references) >
<a id="logicalNetworkMap"></a>
***
###### Logical Network Map
***

These days, of course, may be working with network equipment that isn't physical at all, it may be virtualized - may want to provide a higher level view of how the network connects together. 

In those cases, instead of creating a physical network map, may want to create a logical network map. 

Regardless of the type of map you plan on creating, want to use some specialised software to do this - in Windows, might want to use ``Visio`` - Mac OS has ``OmniGraffle``. 

There are also third party websites (gliffy.com) that allow you to create these graphics inside of the browser itself.

These would be high level views that would show how a wide area network may be connected. 

Or it may show the logical communication of an application as it goes from a web server into middleware and in the database backend. 

Working with a third party, it's common to use this logical network map to show how an application data flow may be working or to see exactly how the network may be laid out across a large area. 

This is a good starting point and then can work down further to find more details in the configuration as the collaboration goes along.

***
###### Star (Switch Networks)
***

There are many different ways to represent a network topology in these diagrams, one network topology type might be the star network. 

Very common network type in today's switch networks.

Have a centralised device like a switch, and all of the devices are connected directly to that switch.

The switch effectively sits in the middle and everybody else connects to it over these Ethernet links. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Ring (Metropolitan Area Network and Wide Area Network)
***

Another common topology type is the ring topology, where all of the devices are connected to each other in a ring form. 

Token ring networks were very popular local area networks, but now no longer very common.

Would see ring networks in metropolitan area networks and wide area networks.

Common to see these ring networks implemented as dual ring networks - if there is a break in the network, it can loop onto itself for fault tolerance and maintain the uptime of the entire ring by providing that loop onto itself. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Mesh (Wide Area Networks)
***

For even additional fault tolerance, may want to consider implementing a mesh network. 

This is one that has many different links to be able to get to the same place.

It might have all of the different locations connected to every other location, or it might be a partial link whereas some locations are connected to other locations.

Would commonly implement a mesh network if looking for something that was redundant and fault tolerant or if you needed to balance load between multiple links. 

Often see mesh being used in wide area networks. 

Since there are so many opportunities to have a link go down, would have redundancy in a way to route around the problem using this mesh network topology. 

< [Table of Contents](#top) | [References](#references) >

***
###### Bus (Controller Area Network)
***

A bus topology is a connection that all other devices connect to.

Very common use of bus technology was in early networks where we would have a single coax cable and connect all devices onto that coax cable so they could communicate to each other. 

This was obviously a very easy network to implement, because you simply used a single cable that everyone would connect to. 

If there was a break in that cable, it would disable the entire network. 

Might find a modern Bus technology is in your automobile.

This is called a Controller Area Network or a CAN bus type of network.

All of the internal devices inside of your car are able to connect to this single bus and communicate between each other.

###### Wireless ``802.11``
***

Although don't use cables to communicate over our wireless networks, there are still some logical network topologies that we use with our ``802.11`` wireless networks.

***
###### Wireless - Infrastructure (Access Point)
***

Infrastructure topology is the one that most people are using when they are communicating with an access point.

Would join an access point network and then you're able to communicate with everyone else who happens to be on that network. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Wireless - Ad hoc (No Access Point in the Middle) - Frequency
***

Do not have to have an access point to be able to communicate with ``802.11``.

Can have devices communicating directly with each others without putting an access point in the middle - call this ad hoc networking. 

Do not need any additional hardware - simply configure both end stations to communicate to each other over this wireless frequency. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Wireless - Mesh (Internet of Things)
***

Today's Internet of Things wireless networks often use mesh networking. 

These are individual devices that are able to discover each other and use each other as a large cloud of devices. 

This allows these devices not only to self form into this mesh cloud, they can also self heal. 

If a device is disabled, they can simply use this large number of devices to route around the problem and enable communication using a different path.
