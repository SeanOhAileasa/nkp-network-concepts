###### Internet of Things Topologies
***

We tend to be connecting more and more devices to each other into what we are calling the internet of things. For example, we might be connecting wearable technology like a smartwatch or a health monitor, and making those devices communicate to each other. These can also track our locations and provide mapping information for us. The real question, of course, is, what type of data are these storing and where is that information being stored. 

We're also connecting a lot of home automation devices as well. Many people have video doorbells on their house that are connected over the internet, along with their garage door openers, you can control your heating and cooling from wherever you happen to be in the world. And all of these devices know when you're home, and they know when you're not home. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Z-wave
***

One category of Internet of Things technologies is Z-Wave. 

This is a technology that's primarily used to automate the home. So if you need to control lights turning on and off, or you need to lock doors or control the garage door, you can do it using Z-Wave technology. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Z-wave - Wireless Mesh Networking
***

These Z-Wave devices communicate through a wireless mesh network.

You would put many different Z-Wave devices in your home, and they would hop through each other on their way to the destination. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Z-wave - 900 MHz Industrial Scientific and Medical (ISN) Band
***

Z-Wave Devices commonly communicate over the 900 MHz Industrial Scientific and Medical (ISN) band.

That's the industrial, scientific and medical band. And there's no special licence required to be able to use those frequencies. And since they're using the 900 MHz band, there's no conflict with ``802.11`` networks that might be running in your home. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ANT/ANT+
***

Another common type of Internet of Things topologies is the ANT or ANT+ wireless network protocol. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ANT/ANT+ - 2.4 GHz Industrial Scientific and Medical (ISN) Band
***

These are using the 2.4 GHz Industrial Scientific and Medical (ISN) band. 

And you commonly see these associated with fitness devices and heart rate monitors. 

These are using a completely different type of networking than your ``802.11`` or your Bluetooth, but they are using 2.4 GHz.  

These frequencies of course could be jammed, and ANT and ANT+ don't require any type of encryption to be able to communicate between all of these devices. 

< [Table of Contents](#top) | [References](#references) >

***
###### Bluetooth - Personal Area Network (PAN)
***

One of the most popular Internet of Things topologies is Bluetooth. 

This is using the Personal Area Network (PAN) to connect a lot of our personal devices together. 

You can use Bluetooth to connect wireless headphones and external speakers or could use it for tethering internet connections from your mobile phone, or it might be connected to help monitoring systems all connected together through Bluetooth. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Near Field Communication (NFC)
***

Another common Internet of Things technology is near field communication or NFC. 

We most commonly see this integrated into our mobile phones. 

It's a two-way communication and it's usually implemented as a way to provide payments using your mobile phone. Simply move your mobile phone close to the payment system and the transaction is done without having to pull out any type of credit card. 

You might also see NFC use with other types of networks. For example, you can use NFC to help with the pairing process for Bluetooth. 

In some environments, you can use NFC as an access token, so you can use your mobile phone to allow you access into a room with an electronic lock. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Infrared
***

Infrared is a technology that's been around for a very long time. 

But we're integrating it also into our mobile phones, our tablets and our smart watches. 

Although in the past, we used infrared for file transfers and printing. 

The most common use these days of infrared is to control an entertainment centre, all by using your mobile device. 

< [Table of Contents](#top) | [References](#references) >

***
###### Radio Frequency Identification (RFID)
***

Radio Frequency Identification, or RFID is another Internet of Things technology. 

And we seem to use RFID everywhere in our daily life. It's used in our access badges in inventory tracking and some of us even have RFID tags inside of our pets. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Radio Frequency Identification (RFID) - Powered using Radio Energy
***

RFID tags are usually not powered devices, they're powered using the radio energy that is sent to the tag. 

It powers up the tag and the tag sends back an ID over this wireless communication. 

This is a bi-directional communication. 

And you can find some RFID tags that do use a powered or active format as well. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### IEEE ``802.11``
***

And one of the most common Internet of Things technologies is our ``802.11`` wireless networks. 

We use them in our home, we use them work and we use them everywhere else as well. 

This is an IEEE standard and it's managed by the IEEE 802. Committee. 

There have been many updates to ``802.11`` Over time, and we foresee many updates coming in the future as well for this very popular wireless technology. 

If you want to be sure that your Wi Fi devices will be able to communicate with each other, you need to look for the Wi Fi trademark on the devices which help ensure that all of these will be able to interoperate with each other.
