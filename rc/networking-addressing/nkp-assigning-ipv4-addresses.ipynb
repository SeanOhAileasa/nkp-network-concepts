***
###### Assigning IPv4 Addresses
***

In the early days of IPv4, we would have to go from workstation to workstation to workstation to manually configure all of the IP configuration details such as the IP address, the subnet mask, the default gateway, DNS servers that were required, NTP information, and anything else related to the overall configuration of IP. 

< [Table of Contents](#top) | [References](#references) >
<a id="networkConceptsAssigningIPv4AddressesBOOTP"></a>
***
###### Bootstrap Protocol - BOOTP
***

Obviously, this manual configuration process did not scale very well.

In October of 1993, BOOTP was created, this is the bootstrap protocol. 

But BOOTP didn't allow us to configure everything we needed for an IP configuration - there were still a number of manual configurations that were required. 

BOOTP also didn't know when a lease was up and an IP address was available again. 

There was no mechanism in place for BOOTP to realise that there were additional IP addresses that could be leased to other devices. 

< [Table of Contents](#top) | [References](#references) >
<a id="dynamicHostConfigurationProtocolDHCP"></a>
***
###### Dynamic Host Configuration Protocol (DHCP)
***

Given limitations of BOOTP, the Dynamic Host Configuration Protocol (DHCP) was created in 1997.

The standard has been updated through the years. 

DHCP allows us to configure most IP configuration settings, and it works across almost any device that you would connect to the network. 

< [Table of Contents](#top) | [References](#references) >
<a id="dynamicHostConfigurationProtocolDHCPDHCPPool"></a>
***
###### DHCP Pool
***

Before our workstation was turned on our DHCP servers were configured with a list of available IP addresses that could be assigned to other devices on the network.

Called a pool of IP addresses, its very common to configure multiple pools of IP addresses for these DHCP servers. 

This pool of available IP addresses will shrink and grow again, based on how many people are on the network and how many may have left the network. 

If you visit a network a second time, your IP address may change depending on what IP addresses are available in that pool. 

< [Table of Contents](#top) | [References](#references) >
<a id="dynamicHostConfigurationProtocolDHCPTurningDynamicintoStatic"></a>
***
###### Turning Dynamic into Static
***

However, there are a number of devices on your network that you might want to always have exactly the same IP address, and don't want them to change based on what might be available in a DHCP pool. 

Might have servers, printers or other infrastructure devices that you would like to administratively assign an IP address, and never want that IP address to change. 

< [Table of Contents](#top) | [References](#references) >
<a id="dynamicHostConfigurationProtocolDHCPTurningDynamicintoStaticManualConfiguration"></a>
***
###### Manual Configuration
***

One way to accomplish this is to manually configure an IP address on every single server printer and infrastructure device. 

This would require disabling DHCP and then manually typing in the IP address, subnet mask, DNS information and anything else that might be required for the IP configurations on that device.

This obviously means that you'll need to log into every server, every printer and every infrastructure device so that you can manually configure these. 

If anything changes with your IP address configurations, you'll have to revisit all of those devices and manually make all of the changes necessary to keep those devices available on the network. 

< [Table of Contents](#top) | [References](#references) >
<a id="dynamicHostConfigurationProtocolDHCPTurningDynamicintoStaticIPReservation"></a>
***
###### IP Reservation - Server MAC Associate IP Address
***

To avoid all of that administration, might want to leave DHCP enabled on those infrastructure devices, then on your DHCP server, you would create an IP reservation that would associate the MAC address of that server with an IP address that you would administer. 

This means that all of the infrastructures have IP addresses that you've assigned. 

If ever need to make any changes, simply make those changes in the DHCP server instead of having to visit every single one of those devices to make the change manually. 

< [Table of Contents](#top) | [References](#references) >
<a id="automaticPrivateIPAddressingAPIPA"></a>
***
###### Automatic Private IP Addressing (APIPA) - Link-Local Address
***

An interesting thing happens if you've configured a workstation for DHCP and no DHCP server is available. 

Although you might think the workstation would not be able to communicate on the network without an IP address, there is a type of IP address that allows the workstation to at least communicate to local devices. 

We call this a link-local address.

Although this device would not be able to communicate through a router, it would still be able to communicate to all other devices on its local IP subnet. 

In IPv4, we call this link-local address, an ``APIPA`` address that stands for Automatic Private IP Addressing.

< [Table of Contents](#top) | [References](#references) >
<a id="automaticPrivateIPAddressingAPIPAReservedAddresses"></a>
***
###### Reserved [``169.254.0.1``,``169.254.255.254``]
***

There's an address range that has been reserved for ``APIPA`` - [``169.254.0.1``,``169.254.255.254``].

```python
169.254.0.1
169.254.255.254
```

< [Table of Contents](#top) | [References](#references) >
<a id="automaticPrivateIPAddressingAPIPAFunctionalAddresses"></a>
***
###### Functional [``169.254.1.0``,``169.254.254.255``]
***

The APIPA standard also reserves the first and last ``256`` address block in this address range so the functional IP addresses that you would see on a device would be ``169.254.1.0`` through ``169.254.254.255``.

So the functional IP addresses that you would see on a device would be ``169.254.1.0`` through ``169.254.254.255``.

```python
169.254.1.0
169.254.254.255
```

This means if this workstation, which normally would get an IP address from DHCP cannot communicate to that server, it will simply pick one of these addresses inside of this range - [``169.254.1.0``,``169.254.254.255``].

This device then sends an ARP request to the network to make sure no one else on this subnet happens to be using this particular IP address, then assigns this address if nobody responds.

<a id="automaticPrivateIPAddressingAPIPAConfiguration"></a>
***
###### Configuration ``169.254``
***

The way that you would tell that a device has been assigned one of these APIPA addresses is to look at the IP address configuration for that operating system. 

If the IPv4 address starts with ``169.254``, you know this device is using an APIPA address.
