###### Early IP Protocol Subnet Masks
***

In the early days of the IP protocol, there were only three subnet masks that could possibly be assigned to a particular workstation:

i. Class A


ii. Class B


iii. Class C

< [Table of Contents](#top) | [References](#references) >
<a id="networkConceptsEarlyIPProtocolSubnetMasksClassA"></a>
***
###### Class A Subnet Mask ``255.0.0.0``
***

The subnet mask for a Class A network is ``255.0.0.0``. 

This also means there are ``8`` bits available for the network address, and ``24`` bits available for hosts on those networks.

< [Table of Contents](#top) | [References](#references) >
<a id="networkConceptsEarlyIPProtocolSubnetMasksClassB"></a>
***
###### Class B Subnet Mask ``255.255.0.0``
***

A Class B subnet mask is ``255.255.0.0``.

There are ``16`` bits available for the network and ``16`` bits available for the hosts.

< [Table of Contents](#top) | [References](#references) >
<a id="networkConceptsEarlyIPProtocolSubnetMasksClassC"></a>
***
###### Class C Subnet Mask ``255.255.255.0``
***

A Class C subnet mask is ``255.255.255.0``, where there are ``24`` bits that are available for the network and ``8`` bits available for hosts.

< [Table of Contents](#top) | [References](#references) >
<a id="ClassBasedSubnetMasks"></a>
***
###### Class-based Subnet Masks
***

Drawing lines through the subnet mask, can see where the separation is between the network address and the host.

This line tends to change, whether it’s a Class A, a Class B, or a Class C subnet mask.

As IP became more popular, we realized that limiting our subnet masks to a Class A, B, or C was not a very efficient way to perform IP addressing. 

So we really haven’t used this method of automatic subnet masking since 1993. 

But we still use these terms in casual conversation. We might have a network that’s a ``192.168.1.1``, and somebody might ask, what’s the subnet mask on this network? And we’ll simply say, it’s a Class C ( much faster way than saying, it’s ``255.255.255.0```). 

The reason we still reference these class-based subnet masks is that they become a starting point if we ever want to provide any additional subnetting to a particular IP address range.

<a id="ClassAIPAddresses"></a>
***
###### Class A IP Addresses 1st Octet [``1``,``126``] (``0``,``127``)
***

Class A subnet masks are associated with IP addresses that begin with the number ``1`` through ``126``, and the subnet mask for Class A is ``255.0.0.0``.

< [Table of Contents](#top) | [References](#references) >
<a id="ClassBIPAddresses"></a>
***
###### Class B IP Addresses 1st Octet [``128``,``191``] (``127``,``192``)
***

If an IP address happens to start with a ``128`` through ``191`` as the first octet, then that is a Class B address, and the default subnet mask for that address is ``255.255.0.0``. 

< [Table of Contents](#top) | [References](#references) >
<a id="ClassCIPAddresses"></a>
***
###### Class C IP Addresses 1st Octet [``192``,``223``] (``191``,``224``)
***

The first octet of a Class C address will start with a ``192`` through ``223``, and the default subnet mask for this Class C address is ``255.255.255.0``.

###### IP Address Associated Class 
***

Need to know these default classes so you know where to start when you begin IP subnetting. 

- ``17.22.90.7`` - refer back to our chart of IP address classes, we know that if the first octet happens to be a value between ``0`` and ``126``, then it is a Class ``A`` address.

- ``220.10.77.40`` - look at that first octet, which is ``220``, know that our Class ``C`` range is between ``192`` and ``223`` - 220 certainly fits inside of that range, so this is a Class ``C`` address.

- ``165.245.0.1`` - want to look at the first octet, which is ``165``, and referring back to the chart above, know that anything between ``128`` through ``191`` is a Class ``B`` address.

- ``128.90.10.2`` - starts with a ``128`` - know that Class ``B`` addresses begin with ``128`` through ``191``, which makes this a Class ``B`` address.

- ``191.77.24.250`` - starts with a ``191`` and again, we know our Class ``B`` range is between ``128`` and ``191``, so this is also a Class ``B`` address.

- ``192.1.12.5`` - know that ``192`` falls into the Class ``C`` subnet range, so this would be a Class ``C`` IP address.

***
###### Subnet Construction
***

Using these IP Addresses and Subnet Masks to start architecting the values that we could use for devices on a particular IP subnet. 

< [Table of Contents](#top) | [References](#references) >
<a id="subnetConstructionNetworkAddress"></a>
***
###### Network Address
***

The first address to calculate is a network address.

This would be the first IP address that happens to belong to a particular subnet. 

Can also get this value by taking all of the host bits of a particular subnet and changing them all to zero, and that will allow you to calculate the network address. 

< [Table of Contents](#top) | [References](#references) >
<a id="subnetConstructionHostAddress"></a>
***
###### Host Address
***

If the very first IP address on a particular IP subnet is the network address, then one that is just one number above that is the first usable host address. 

If the the network address is calculated, it’s very easy to find the first usable host address - just add one to the IP address.

Conversely, can find the network address by determining the first usable host address and subtracting one from that. 

< [Table of Contents](#top) | [References](#references) >
<a id="subnetConstructionBroadcastAddress"></a>
***
###### Broadcast Address
***

The last IP address on a particular IP subnet is the network broadcast address. 

Setting all host bits to 1, this would allow you to calculate the network broadcast address.

Want to define the last usable IP address that you can assign to a workstation or some other device on that subnet, you would find the network broadcast address and the last usable host address would be one number lower than that broadcast address.

< [Table of Contents](#top) | [References](#references) >
<a id="subnetCalculations"></a>
***
###### Subnet Calculations ``10.74.222.11`` Class A ``255.0.0.0``
***

Using our knowledge of the network address, the first usable IP address, the broadcast address, and the last usable IP address to do some calculations using an actual IP values.

When performing this calculation, let’s assume that we’re using the default values or the class-based subnet masks.

Know that anything that starts with an IP address of ``10`` is within the Class ``A`` subnet range, which means the subnet mask will be ``255.0.0.0``.

Knowing the subnet mask, now able to draw that line that will separate out the network part of the IP address versus the host part of the IP address.

< [Table of Contents](#top) | [References](#references) >
<a id="subnetCalculationsNetworkAddress"></a>
***
###### Network Address ``10.0.0.0``
***

We know our network value is everything to the left of the line, so that would be ``10``, and then our host values would be everything to the right of the line, which would be ``74.222.11``.

To find the network address, then, we set all of the host bits to ``0``.

So instead of ``74.222.11``, we have ``0.0.0``, which means the network address is ``10.0.0.0`` for this particular IP subnet.

< [Table of Contents](#top) | [References](#references) >
<a id="subnetCalculationsFirstHostAddress"></a>
***
###### ``[Host Address,`` ``10.0.0.1``
***

To find the first available host on the subnet, we add ``1`` to the network address, which means the first available host could use an IP address of ``10.0.0.1``.

< [Table of Contents](#top) | [References](#references) >
<a id="subnetCalculationsBroadcastAddress"></a>
***
###### Broadcast Address ``10.255.255.255``
***

To find the broadcast address for this network, we would change all of these host bits to be ``1`` in binary, which means that our broadcast address would be ``10.255.255.255``.

````python
network=1
2**8-network
````

< [Table of Contents](#top) | [References](#references) >
<a id="subnetCalculationsLastHostAddress"></a>
***
###### ``,Host Address]`` ``10.255.255.254``
***

To find the last available host that you could use on this IP subnet, we subtract ``1`` from the broadcast address, making the last available IP address on the subnet ``10.255.255.254``.

````python
network=1
2**8-network-1
````
