***
###### The Subnet Mask
***

May see a subnet mask written in a number of different ways.

Demonstrating how to convert between all of these different representations. 

Can think of the subnet mask, though, as a contiguous series of ones, where the ones are on the left side of the subnet mask, and the zeros are on the right side of the subnet mask.

So if we have a subnet mask, that has three octets of all ones, we know that this subnet mask is ``255.255.255.0``.

< [Table of Contents](#top) | [References](#references) >
<a id="classlessInnerDomainRoutingCIDR"></a>
***
###### Classless Inner Domain Routing - CIDR
***

Might also see this subnet mask written as a ``/24``. There are ``24`` bits in this particular subnet mask, in the first three octets. Also called CIDR block notation - C-I-D-R that stands for Classless Inner Domain Routing.

Might also hear this referred to as a prefix notation or slash notation because of the use of the ``/24`` to specify the number of bits - the slash is used to designate the number of bits that will be used for the subnet mask.

So for a subnet mask of ``/24``, know that that is the size of the network and the remaining bits, which would be ``8`` bits, in this case, would be the number of bits available for hosts.

< [Table of Contents](#top) | [References](#references) >
<a id="binarytoCIDRblocknotation"></a>
***
###### Binary to CIDR-block notation
***

Performing some conversions between the binary representation of the subnet mask and the CIDR block or the slash type notation of the subnet mask.

Have a subnet mask written out in binary, which has all ones in the first two octets and all zeros in the last two octets.

Going to add up the number of ones in this entire subnet mask, which means that we would have ``8 + 8 + 0 + 0``, which means there is ``/16``.

There is ``16`` bits in this subnet mask. 

That means that our network is going to be ``16`` bits long, and the host values are also going to be ``16`` bits long. 

Performing the same conversion with this subnet mask, which is all ones in the first three octets, two ones in the last octet, and of course, zeros after that.

Adding all of those up, ``8 + 8 + 8 + 2``, we have a total of ``26`` for a ``/26`` subnet mask.

This means that the network is going to be ``26`` bits, and the host value will use the ``6`` bits at the end.

Another binary subnet mask, which is all ones in the first octet, four ones and four zeros in the second octet, and then all zeros after that. 

Add up all of the ones, which is ``8 + 4 + 0 + 0`` for a ``/12`` subnet mask. 

Gives ``12`` bits to use for network addresses and ``20`` bits to use for hosts on each one of those networks.

< [Table of Contents](#top) | [References](#references) >
<a id="binarytoDecimal"></a>
***
###### Binary to Decimal
***

If you only have ``8`` bits available in each one of these octets, there’s only a certain number of combinations of these binary to decimal representations you can have.

All zeros would obviously be ``0``.

A ``1`` and all ``0`` is going to be ``128``. 

Two ones and all zeros is ``192`` and so on.

So instead of performing a manual calculation every time you wanted to convert between binary and decimal, you could simply write out this chart and very easily be able to convert between them.

< [Table of Contents](#top) | [References](#references) >
<a id="binarytoCIDRblockNotation"></a>
***
###### Binary to CIDR Notation
***

The subnet mask with all ones in the first octet and four ones in the second octet and zeros after that.

Looking at the chart, all ones is ``255``.  Then four ones and four zeros, is ``240``. So the subnet mask would be ``255.240.0.0``.

There are ``12`` ones, so the subnet mask will be a ``/12`` CIDR block address, giving ``12`` bits for the network and ``20`` bits for all of the hosts that would be on each network. 

Using the chart to perform the same calculation with this subnet mask. 

In binary, it’s all ones in the first octet, all ones in the second octet, and three ones in the third octet.

Know that all ones is going to be ``255``, and three ones and the rest zero would be ``224``, making the subnet mask ``255.255.224.0``.

Adding up the number of ones, there are ``/19`` for this subnet mask, giving us ``19`` bits for the network and ``13`` bits to use for hosts on each of those networks.

<a id="cIDRNotationtoBinary"></a>
***
###### CIDR Notation to Binary
***

Let’s do this in reverse, where we have the CIDR block notation, and we want to convert back the decimal representation of that.

Have a ``/26`` as our subnet mask.

Writing this out manually, in decimal form, there are ``26`` ones and then the remaining zeros for the subnet mask. 

Looking at the reference chart, we know that all ones is ``255``, and then two ones at the beginning would be ``192``, making this subnet mask ``255.255.255.192``. 

There is ``26`` bits for the network address and ``6`` bits for the host on each network.
