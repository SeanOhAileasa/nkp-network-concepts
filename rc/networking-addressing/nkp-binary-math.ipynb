###### Binary Math
***

There are only two numbers to know when dealing with binary - ``0`` and ``1``.

The single digit is a bit in the binary system and putting eight bits together, refer to that as a byte (referred to as an octet to designate that there are indeed eight bits for every byte).

This binary to decimal conversion chart starts at the far right with the number 1, and each column is doubled all the way up to 128. This is the conversion chart to use to be able to convert from binary to decimal and back again.

The binary conversion chart created has eight separate columns and thatâ€™s because those eight columns group together to form a single byte or a single octet.

Could extend these columns so that theyâ€™re long enough to perform whatever binary calculation is needed. 

But almost everything referring to binary calculations revolves around IP addressing, which means there will be groupings of 8 bits in a single byte.

###### Binary to Decimal Conversion
***

Using this conversion chart to convert from a binary value to a decimal value. 

The first thing to do is write out all of those individual bits in order and of course, trying to find what the decimal value of that might be. 

Just above these bits, weâ€™re going to write our conversion chart.

Start with 1, double it to 2, to 4, 8, all the way to 128. 

Can see there is a separate value listed along with each individual bit that weâ€™re trying to convert. 

Next add all of these values based on the numbers in the conversion chart. 

Every place there is is 0, weâ€™re going to bring down a 0. But every place there is a 1, weâ€™re going to bring down the value thatâ€™s associated with that place in our conversion chart.

Just above the 1 is the number 2 therefore bring down a 2.

Put all of these together, add them all up, will get a value of 2 decimal.

Got binary ``1 1 1 1 1 1 1 1`` so bringing all 1s down and putting all of the conversion chart numbers at the top. 

This means pulling down every single 1 of those conversion chart numbers, add them all together to come up with ``255`` in decimal.

###### Decimal to Binary Conversion
***

Reverse the process.

Start with a decimal value, and letâ€™s determine what the binary equivalent is of that decimal value. Decimal ``154``.

Normally, would list out the binary numbers, but obviously, do not know what those are yet. 

Will put the conversion chart right on the top of these unknown values. 

There is only one combination of 0s and 1s that would equal ``154`` in binary.

Step through each one of those decisions to see if we put a ``1`` or ``0`` in each one of those columns.

Start with the ``128`` that happens to be on the far left, and ask, is ``128`` less than or equal to the ``154`` decimal.

In this case, ``128`` is less than or equal to ``154``, quite a bit less therefore putting a 1 in that particular place, and weâ€™ll bring down the ``128``. 

Looking at the next value, which is a ``64``.

Adding ``128`` and ``64``, which happens to be ``192`` - ask is that less than or equal to ``154``? In this case, no. 

Itâ€™s more than ``154`` so would put a ``0`` in that particular column.

Letâ€™s perform the same question with the next column over, which is the 32. Is ``128`` plus ``32``, which is ``160``, less than or equal to ``154``? 

That is not less than or equal to ``154`` so we put a ``0``.

The next values a ``16``. 

Is ``128`` plus ``16``, which is ``144``, less than or equal to ``154``?

It is less than or equal to that value so putting a ``16`` down. Now, ``144`` is the number that we have so far, ``128`` plus ``16``.

So is ``144`` plus ``8``, our next column, less than or equal to ``154``? 

It is, ``152`` is less than or equal to ``154``. 

Bring down the value of ``8``. 

Ask the same question of the next column, which is a ``4``.

Is ``152`` plus ``4`` less than or equal to ``154``? 

Itâ€™s not. Itâ€™s more than ``154``. 

Ask the same question in the ``2`` column. 

Is ``152`` plus ``2``, which is ``154``, less than or equal to ``154``?

And it is equal to ``154``.

And now that we finally reached that value, all of the remaining bits will happen to be ``0``.

Thereâ€™s only one column left, therefore putting a ``0`` in that column.

The binary representation of ``154`` decimal is ``1 0 0 1 1 0 1 0``.

###### More Bits then More Addresses
***

If there are a fewer number of bits available for a calculation, then that means there are a fewer number of decimal representations of that value. 

For example, if there are two bits available, the only possible options are ``0, 1, 2, and 3``.

If thereâ€™s three bits available, got ``0 through 7``.

If four bits are available, got ``0 through 15``, and so on.
