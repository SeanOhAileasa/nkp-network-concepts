###### Unicast - One-to-One Relationship - Web Browsing Session 
***

In network communication a unicast is when one device is sending information to one other device.

It's a one-to-one relationship. 

That's why we call it a unicast, since there is a single station, communicating with one other station, so there's always two devices that are involved in this unicast communication.

It's common to see unicast when doing things like performing a web browsing session, your single device will be accessing a single web server or if you're transferring a file from your machine to another machine, that's commonly using a unicast communication. 

There are many applications where using a unicast may not be the most efficient way to transfer data. Take an example of a live event that you want to stream to multiple devices. To stream this real-time information, you need to set up a session with every single device that wants to receive that data. So the more devices that were viewing that real-time data over that unicast communication would increase the amount of bandwidth every time you started to use another unicast session.

###### Broadcast - All - Routing Updates & OS Communication - IPv4 Address Resolution Protocol ``ARP``
***

A broadcast is the opposite of a unicast. 

One device is communicating to every other device on the network, all at the same time. 

One device would send one packet and that single packet would be received by everyone else that was on that local network. 

There is a limited scope to these broadcasts, that's why we have a broadcast domain, that limits the number of devices that might receive a broadcast put onto a local subnet. 

You commonly see broadcast use for things like routing updates and operating system communication.

Its common to see broadcasts used with IPv4 when an ``ARP`` request is made across the network. 

But broadcast frames could be problematic - if there are too many broadcasts and every device has to gather every broadcast frame, it could slow down the overall performance of the network. 

That's why IPv6 was specifically designed not to use broadcasts at all, but to instead focus on using multicast instead.

###### Multicast - One-to-Many Relationship - Multimedia Delivery
***

You can think of a multicast as a compromise between a unicast and a broadcast, we are sending to multiple devices on the network, but we're not sending to every device on the network. 

Think of it as a one-to-many communication. 

This is commonly used for things like multimedia delivery, where you can send out a single stream of information and anyone who's interested in receiving that stream of data can subscribe or connect to that particular multicast. 

Multicast have a limited scope and it sometimes is difficult to scale them in very large environments. All of your infrastructure devices have to understand how to deal with multicast and all of your in devices have to know how to subscribe or view that multicast information.
