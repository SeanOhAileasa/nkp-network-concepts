###### Cellular Network Standards
***

We rely on cellular networks to provide most of our mobile communications these days.

We use our mobile phones and mobile devices to communicate over the separate cells of communication. 

We have antennas that can converge the data within the cells, so that you can move from cell to cell and still maintain communication wherever you happen to be. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### 2G Networks
***

In the early days of cellular networking, there were two primary standards that you would see:

i. GSM, which is the Global System for Mobile Communications. 

ii. CDMA or Code Division Multiple Access. 

These were really designed for voice communication.

There is limited data functionality.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### 2G Networks - Circuit Switching
***

It used circuit switching to be able to set up a call send data, and it had to emulate that same circuit switching to be able to send data.

To be able to perform some type of data or packet switching, we needed newer standards that would support these newer ways to communicate. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### 2G Networks - i. Global System for Mobile Communications (GSM) 
***

The Global System for Mobile Communications (GSM), had about a 90% coverage in the market. That's because it was a standard almost everywhere in the world.

And the United States, AT&T and T-Mobile were big providers of the GSM standard. 

This also allowed you to take a SIM card out of one phone and move it to another phone. So anytime you needed to change telephones, you could simply move your SIM card from phone to phone.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### GSM - Time Division Multiple Access (TDMA) Multiplexing
***

This original GSM standard uses a multiplexing called TDMA to be able to communicate where everyone on a particular frequency gets a little slice of time to be able to complete that communication. 

Here's how Time Division Multiple Access or TDMA multiplexing works.

i. You would have multiple streams, you can see each of these phones on the left side communicating over a blue, a red and a purple communication. 

ii. The multiplexer blends those together and gives everybody a certain time stamp that's available to them. 

iii. On the other side, that signal is demultiplexed - would break out the blue, the red and the purple streams so that they can be sent to their normal destinations.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### 2G Networks - ii. Code Division Multiple Access (CDMA)
***

The other standard of the time was CDMA - Code Division Multiple Access, which uses a different method of multiplexing. 

This data with CDMA, everyone's communicating over the same frequencies, but each call is using a different code. And the phone on the other end is filtering out all of the information that it doesn't need to hear.

This was a standard commonly used by Verizon and Sprint. And all of those handsets were controlled by that network provider. In the United States, there wasn't much adoption outside of Verizon and Sprint for CDMA. 

< [Table of Contents](#top) | [References](#references) >

***
###### 4G Networks - Long Term Evolution (LTE)
***

These days, most mobile phone providers have moved towards what they call 4G or the fourth generation of Long Term Evolution communication or LTE. 

This is a converged standard, so we no longer have to worry about a provider giving us a GSM or a CDMA phone.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### LTE - GSM and EDGE (Enhanced Data Rates for GSM Evolution)
***

This is based on the GSM and EDGE which is Enhanced Data Rates for GSM Evolution, and it finally allows us to completely and seamlessly communicate with voice and data simultaneously over the same network.

###### LTE - Download Data Rates up to 150 Mbits/s
***

The original LTE standard supported download data rates up to 150 Mbits/s (megabits per second).

###### LTE - LTE Advanced - Download Data Rates up to 300 Mbits/s
***

The latest version of LTE - LTE Advanced has doubled that to support download speeds of up to 300 megabits per second.
