###### ``802.11`` Wireless Standards
***

Almost all of us have devices that use ``802.11`` wireless networks.

It might be our mobile phones, it might be a tablet, or it might be the computer or laptop that we're using. 

Most of us rely every day on this ``802.11`` standard that comes from the IEEE 802 Standards Committee. This is one of the most popular wireless technologies in the world. Going to step through the evolution of ``802.11`` wireless networking through the years. 

When the ``802.11`` standards were released, there were two standards that were released simultaneously. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11a``
***

One of these is the ``802.11a`` standard - one of the original standards that was released in October of 1999.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11a`` - 5 GHz Frequency Range - Throughput up to 54 Mbit/s
***

``802.11a`` operated in the 5 GHz range and gave us throughputs up to 54 megabits per second (Mbit/s). 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11a`` - Range: Absorbed by Objects (5 GHz)
***

Because ``802.11a`` was using 5 GHz frequencies, it tended to have a smaller available range than the ``802.11b`` standard.

``802.11b`` uses 2.4 GHz frequencies, and those lower frequencies tend to bounce off of other objects, rather than being absorbed by objects at higher frequencies. 

The ``802.11a`` standard has been replaced by newer ``802.11`` standards, so you don't tend to see much ``802.11a`` traffic on a network today. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11b``
***

The ``802.11b`` standard was released simultaneously with the ``802.11a`` standard in October of 1999. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11b`` - 2.4 GHz Frequency Range - Throughput up to 11 Mbits/s
***

As we've mentioned, it uses the 2.4 GHz range to communicate.

``802.11b`` communicated at a maximum possible throughput of 11 megabits per second (Mbits/s), which was significantly different than the 54 megabits per second (Mbits/s) available with ``802.11a``.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11b`` - Range: Bounce off of Objects (2.4 GHz)
***

But ``802.11b`` gave us effectively longer ranges, because those 2.4 GHz frequencies would bounce off of objects rather than being absorbed.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11b`` - Conflicts (2.4 GHz)
***

When challenged with ``802.11b`` and these 2.4 GHz frequencies however, is there are a lot of other devices that are communicating at that frequency - cordless phones, baby monitors and Bluetooth communications all operate at 2.4 GHz and can potentially create conflicts with the ``802.11b`` communication.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11g`` - Upgrade to ``802.11b``
***

``802.11g`` was introduced in June 2003. 

It was very similar to ``802.11b`` and this is considered an upgrade to ``802.11b``. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11g`` - 2.4 GHz Frequency Range - Throughput up to 54 Mbits/s
***

It operates in the same 2.4 GHz range. 

But the overall throughput was increased to 54 megabits per second (Mbits/s), which is very similar to the throughput for ``802.11a``. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11g`` - Backwards Compatible with ``802.11b``
***

This ``802.11g`` standard was backwards compatible with ``802.11b``. So if you installed an ``802.11g`` access point, it could very easily also accommodate ``802.11b`` clients. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11g`` - Conflicts (2.4 GHz)
***

But you do have the same problem with frequency conflict because you're still communicating in the 2.4 GHz frequency range. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11n``
***

And October of 2009, we received a big update to the ``802.11`` standard - this was ``802.11n``.

This was effectively an update for ``802.11g`` and ``802.11b`` and ``802.11a``.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11n`` - 5 GHz or 2.4 GHz Frequency Range - Channel Widths of 40 MHz - Throughput up to 600 Mbit/s
***

``802.11n`` can operate at 5 GHz or 2.4 GHz. And these particular channel widths of 40 MHz are much larger than the bandwidth available with previous versions of ``802.11``.

``802.11n`` can support throughputs up to 600 megabits per second (Mbits/s), using this 40 megahertz channel width, and four antennas sending multiple streams of data simultaneously. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11n`` - Multiple Input Multiple Output (MIMO)
***

``802.11n`` was the first version of ``802.11`` to use multiple input multiple output or MIMO, where you're able to send multiple streams of information over the same frequency. You just need multiple antennas and radios to be able to send that data. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11ac``
***

One of the most recent versions of ``802.11`` is ``802.11ac``. 

This was introduced in January of 2014.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11ac``  - 5 GHz Frequency Range - Router 2.4 GHz Communicates using ``802.11n``
***

``802.11ac`` operates in the 5 GHz band exclusively. 

There are some ``802.11ac`` routers that you'll see communicating in both 5 GHz and 2.4 GHz - but all of the 2.4 GHz communication is communicating using ``802.11n``. 

Because there's no requirement to communicate in the very crowded 2.4 GHz band, you can use much larger bandwidths within 5 GHz and ``802.11ac`` supports up to 160 MHz channel bandwidth. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11ac`` -  Bond Creating Larger Channel Bandwidths
***

This ``ac`` version of ``802.11`` can also bond together individual channels to effectively create larger channel bandwidths. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### ``802.11ac`` -  Denser Signalling Modulation - 8 Multi-User Mulitiple Input Multiple Output (MU-MIMO)
***

There's also a denser signalling modulation which means you can send data much faster with eight multiuser MU-MIMO streams on ``802.11ac`` you can theoretically support almost seven gigabits per second of throughput on this wireless network. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Summary ``802.11`` Networking
***

Here's a summary of these five standards ``802.11a`` and ``802.11b`` that were released simultaneously:

- ``a`` supported 5 GHz and ``b`` supported 2.4 GHz.


- Neither of them supported multiple input multiple output (MIMO).


- ``802.11a`` provided 54 Mbits/s of maximum theoretical throughput and ``802.11b`` supported 11 Mbits/s.

- ``802.11g`` was an update to ``802.11b`` and it uses the same frequencies of 2.4 GHz.


- Not supporting multiple input multiple output (MIMO).


- It increased the speed up to 54 megabits per second (Mbits/s) 

- ``802.11n`` was the very first introduction of multiple input multiple output (MIMO).


- Can support up to four streams in ``802.11n`` running at either 5 GHz or 2.4 GHz.


- Running up to 4 of those 150 megabit MIMO streams, can support a theoretical maximum of 600 megabits per second (Mbits/s) with ``802.11n`` 

- With ``802.11ac`` running in the 5 GHz range.


- Running 8 multiuser MU-MIMO streams, can support almost seven gigabits of throughput over this wireless network.
